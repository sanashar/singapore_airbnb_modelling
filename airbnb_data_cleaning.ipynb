# Data Cleaning - Singapore Airbnb Dataset 

## Loading the data & Data info -

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the downloaded file

df = pd.read_csv("listings.csv", na_values=' ')
df
```

```python
# Display the number of rows & columns

df.shape
```

```python
# Display the dataframe info()

df.info()
```

```python
# Display the missing values

df.isna().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
# Drop all NaN

df = df.dropna()
df = df.reset_index(drop = True)
```

```python
# Display the number of rows & columns left after dropping the NaN

df.shape
```
